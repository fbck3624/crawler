```python
import requests
from bs4 import BeautifulSoup
r=requests.get('https://www.ptt.cc/bbs/NBA/index.html')
r.text
```

```python
soup=BeautifulSoup(r.text,"lxml")
soup
```

```python
for d in soup.find_all(class_="r-ent"):
    print('標題： '+ d.find(class_='title').text)
    print('作者： '+ d.find(class_='meta').find(class_='author').text) #meta可省略!
    print('時間： '+ d.find(class_='meta').find(class_='date').text) #meta可省略!
```

```python
for d in soup.find_all(class_="r-ent"):
    print('標題： ', d.find(class_='title').text.replace('\t', '').replace('\n', ''))
    print('作者： ', d.find(class_='author').text.replace('\t', '').replace('\n', ''))
    print('時間： ', d.find(class_='date').text.replace('\t', '').replace('\n', ''))
```

```python
# 因為原始的資料難以判斷「新/舊」，因此我們必須進入內頁抓取詳細的時間，最終存成 List of Dict 的形式再自行排序。

posts = []

for d in soup.find_all(class_="title"):
   # print(('https://www.ptt.cc'+d.find('a')['href']))    ->原本在這  會出錯
    try:
        post = {}
        post['標題'] = d.text.replace('\t', '').replace('\n', '')
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "lxml")
        post['作者'] = r.find(class_='article-meta-value').text  #用find取'article-meta-value'的第一個
        post['時間'] = r.find_all(class_='article-meta-value')[-1].text #用find_all取'article-meta-value'的最後一個
        posts.append(post)
    except:
        continue
    print(('https://www.ptt.cc'+d.find('a')['href']))    #->換到這邊就沒問題了，例外處理，非每一個title底下都有該網址!
posts = sorted(posts, key= lambda x: x['時間'])
# 補充：List of Dict 的排序方法
# https://stackoverflow.com/questions/72899/how-do-i-sort-a-list-of-dictionaries-by-a-value-of-the-dictionary
posts[-1]
```
