```python
r_pop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data_pop = json.loads(r_pop.text)

r_notPop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data_notPop = json.loads(r_notPop.text)

po_avg_co=0
po_avg_like=0
nopo_avg_co=0
nopo_avg_like=0

#ç†±é–€æ–‡ç« 
count=0
for d in data_pop:
    po_avg_co += d['commentCount']
    po_avg_like += d['likeCount']
    count=count+1
    #print(d['commentCount'])
po_avg_co=po_avg_co/count
po_avg_like=po_avg_like/count
print('ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸:'+str(round(po_avg_co)))
print('ç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸:'+str(round(po_avg_like)))
#éç†±é–€
notcount=0
for d in data_notPop:
    nopo_avg_co += d['commentCount']
    nopo_avg_like += d['likeCount']
    notcount=notcount+1
nopo_avg_co=nopo_avg_co/notcount
nopo_avg_like=nopo_avg_like/notcount
print('éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸:'+str(round(nopo_avg_co)))
print('éç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸:'+str(round(nopo_avg_like)))
```
