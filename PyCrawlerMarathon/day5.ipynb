```python
import requests
# 引入函式庫
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
# 想要爬資料的⽬目標網址
response = r.text
# 模擬發送請求的動作
response
```

```python
import json
data = json.loads(response) 
count=0
# 利利⽤用 json 解析網⾴頁內容
for d in data[0]:    
    print(d) 
    count=count+1
# 取出每⼀一篇⽂文章的標題
print ("共"+str(count)+"筆")
```

```python
for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
    print('')
```

```python
r_pop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data_pop = json.loads(r_pop.text)

r_notPop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data_notPop = json.loads(r_notPop.text)

po_avg_co=0
po_avg_like=0
nopo_avg_co=0
nopo_avg_like=0

#熱門文章
count=0
for d in data_pop:
    po_avg_co += d['commentCount']
    po_avg_like += d['likeCount']
    count=count+1
    #print(d['commentCount'])
po_avg_co=po_avg_co/count
po_avg_like=po_avg_like/count
print('熱門文章平均留言人數:'+str(round(po_avg_co)))
print('熱門文章平均按讚人數:'+str(round(po_avg_like)))
#非熱門
notcount=0
for d in data_notPop:
    nopo_avg_co += d['commentCount']
    nopo_avg_like += d['likeCount']
    notcount=notcount+1
nopo_avg_co=nopo_avg_co/notcount
nopo_avg_like=nopo_avg_like/notcount
print('非熱門文章平均留言人數:'+str(round(nopo_avg_co)))
print('非熱門文章平均按讚人數:'+str(round(nopo_avg_like)))
```
