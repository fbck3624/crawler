```python
import requests
from bs4 import BeautifulSoup
r = requests.get("https://www.books.com.tw/")
r.text
```

```python
soup=BeautifulSoup(r.text,"html5lib")
soup
```

```python
buy=[]
for d in soup.find(class_='nav_box').find(class_='menu style01').find_all('h3'):
    #print(d.find('a').text)
    #print(d.find('a')['href'])
    book=d.find('a').text
    www=d.find('a')['href'].split('=')
    #print(www[1])
    buy.append({
        '類型':book,
        '編號':www[1]
    })
import pandas as pd
pd.DataFrame(buy)
```
